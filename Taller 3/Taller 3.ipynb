```python
import pandas as pd

def menu():
  print('-----BIENVENIDO-----')
  print('1) Fibonacci')
  print('2) Sumatoria')
  print('3) Notas')
  print('4) Fallas')
  print('5) Salir')
  return int(input())

def fib(n):
  if n < 2:
      return n
  else:        
      return fib(n-1) + fib(n-2)

def sumat():
  # Documento: 1018506713 ultimo digito 3 + 2 = 5 --> a=5
  s=0
  for i in range(int(input('Digite hasta cual numero desea hacer la sumatoria: '))):    
    s=s+(3*(1/5)**(i+1))
  print('Sumatoria:',s)

def notas():  
  notas=pd.read_csv('/content/drive/MyDrive/Clases/Metodos Numericos/corte 2/Taller 3/Notas.csv')  
  row=['Estudiante '+str(i+1)+':' for i in range(len(notas))]
  notas.index=row   
  defi=(notas.iloc[:,0:3].mean(1)*0.6) + (notas.iloc[:,3]*0.4)
  notas["Definitiva"]=defi
  print(notas)
  print("Media definitivas: ",round(defi.mean(),2))
  print("Desviacion definitivas: ",round(defi.std(),2))


def fallas():
  fallas=pd.read_csv('/content/drive/MyDrive/Clases/Metodos Numericos/corte 2/Taller 3/Fallas.csv')
  row=['Pepito '+str(i+1)+':' for i in range(len(fallas))]
  fallas.index=row
  noApru=[]
  nombreEst=[]
  porcFallas=[]
  totalHorasClase=fallas.shape[1]*2
  porcPerdi=totalHorasClase*0.2
  for i in range(len(fallas.iloc[:])):
    s=0
    for j in range(len(fallas.iloc[i])):
      s=s+fallas.iloc[i,j]
    if s>=porcPerdi:
      nombreEst.append('Pepito '+str(i+1))
      noApru.append(s)
      porcFallas.append(str(round((100*s)/totalHorasClase,2))+'%')    
  fallasPer=pd.DataFrame()
  fallasPer['Nombre']=nombreEst
  fallasPer['# Fallas']=noApru
  fallasPer['% Fallas']=porcFallas
  print("---Se pierde con un 20% de fallas o 12.4 fallas---\n",fallasPer)

bandera=False
while bandera==False:
  opc=menu()
  try:
    if opc==1:
      i=int(input('Ingrese hasta cual termino desea calcular la serie: '))
      r=fib(i)
      print(r)
    elif opc==2:
      sumat()
    elif opc==3:
      notas()
    elif opc==4:
      fallas()
    elif opc==5:    
      b=False
      while b==False:
        s=str(input('¿desea salir? si/no: ')).lower()      
        if s=='si'or s=='s':
          print('Salir')
          b=True
          bandera=True
        elif s=='no' or s=='n':
          b=True
          bandera=False
        else:        
          print("!Esta opcion no existe¡")
    else:
      print("!Esta opcion no existe¡")
  except ValueError:
    print("!Esta opcion no existe¡")
```
