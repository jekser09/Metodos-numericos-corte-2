```python
# Ejecutar primero esta celda para importar lo necesario
import sympy
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
from IPython.display import display

x,y,b,a =sympy.symbols('x,y,b,a')
```

```python
#Punto 1
#Factorizar
print('a) factorizar')
exp1=27*x+x**4
exp2=18*b-6*b**2+2*b**3
exp3=a*x**2+9*a*x+20*a
display('Factorizar',exp1)
display('Factorizado',sympy.factor(exp1))
display("Factorizar",exp2)
display('Factorizado',sympy.factor(exp2))
display("Factorizar",exp3)
display('Factorizado',sympy.factor(exp3))
```

```python
# Simplificar
print('b) simplificar')
exp4=(x**2-5*x)/(x**2-4*x-5)
exp5=(4*x**2-9*y**2)/(4*x**2+6*x*y)
exp6=(x**4-14*x**2-51)/(x**4-2*x**2-15)
exp7=(a**2*x**2-16*a**2)/(a*x**2+9*a*x+20*a)
display('Simplificar',exp4)
display('Simplificado',sympy.simplify(exp4))
display('Simplificar',exp5)
display('Simplificado',sympy.simplify(exp5))
display('Simplificar',exp6)
display('Simplificado',sympy.simplify(exp6))
display('Simplificar',exp7)
display('Simplificado',sympy.simplify(exp7))
```

```python
# Reducir a su minima expresion
exp8=((1/x)-(2/x**2)-(3/x**3))/((9/x)-x)
exp9=1/(a-(a**2-1)/(a+1/(a-1)))
exp10=x**2/(1-(1/(x**2+((1/x)/(x+1/x))))) + (x**2-2)/(1-(1/(x**2+((1/x)/(x+1/x)))))
display('Reducir a su minima expresion',exp8)
display('',sympy.together(exp8))
display('Reducir a su minima expresion',exp9)
display('',sympy.together(exp9))
display('Reducir a su minima expresion',exp10)
display('',sympy.together(exp10))
```

```python
# Descomponer en fracciones parciales
exp11=x/(x**2+x-2)
exp12=(x**4+1)/(x**5+4*x**3)
exp13=(x**6+1)/(x**6+x**3)
display('Fracciones parciales',exp11)
display('',sympy.apart(exp11))
display('Fracciones parciales',exp12)
display('',sympy.apart(exp12))
display('Fracciones parciales',exp13)
display('',sympy.apart(exp13))
```

```python
# Derivar
exp14=sympy.log((2*x+1)**5/sympy.sqrt(y**2+1))
exp15=(y**2+2)**2*(x**4+4)**4
exp16=sympy.tan(x)**(1/x)
display("f(x)=",exp14)
display('Derivada',exp14.diff(x))
print('')
display("g(y)=",exp15)
display('Derivada',exp15.diff(y))
print('')
display("y=",exp16)
exp16=exp16.diff(x)
display('Derivada',exp16)
display('x=PI/4',exp16.subs(x,sympy.pi/4))
```

```python
# Integrar
exp17=(x-1)/(x**2-4*x-5)
exp18=sympy.log(x)/(x*sympy.sqrt(1+sympy.log(x)**2))
exp19=(1+sympy.log(x))*sympy.sqrt(1+(x+sympy.log(x))**2)
display("Funcion:",exp17)
display("Integral de 0 hasta 4",sympy.integrate(exp17,(x,0,4)))
display("Funcion:",exp18)
display("Integral",sympy.integrate(exp18))
display("Funcion:",exp19)
display("Integral",sympy.integrate(exp19))
```

```python
# Punto 2
#a)
x2=np.linspace(-5,5,100)
f1=np.sin(x2)*np.sin(10*x2)
fig2,ax2 = plt.subplots()
ax2.plot(x2,f1,color="#0008FF",label=r"$f(x)=sin(x)sin(10x)$",lw=3,linestyle=":")
ax2.set_title("Grafica")
ax2.legend(loc="best")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.grid()
```

```python
#b)
x1=np.linspace(-5,5,100)
y1=np.exp(x1)*np.sin(x1)
y1p=np.exp(x1)*np.sin(x1)+np.cos(x1)*np.exp(x1)
y1pp=2*np.exp(x1)*np.cos(x1)
fig,ax = plt.subplots()
ax.plot(x1,y1,color="#2DBC7F",label=r"$y=e^x sin(x)$",linestyle="--",lw=2)
ax.plot(x1,y1p,color="#FF5733",label=r"$y'=e^x sin(x) + cos(x)e^x$",linestyle="--",lw=2)
ax.plot(x1,y1pp,color="#29BAE1",label=r"$y''=2e^xcos(x)$",linestyle="--",lw=2)
ax.legend(loc='best')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid()
```
