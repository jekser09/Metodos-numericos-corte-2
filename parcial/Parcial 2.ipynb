```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
#Cedula: 1018506713 p=3
x =sp.Symbol('x')
p=3
```

```python
#Factorizar:
exp1=p*x**2+9*p*x+20*p
display("Expresión",exp1)
display("Factorizar",sp.factor(exp1))
```

```python
#Reducir a su minima expresión
exp2=(p*x**2)/(p+p/(x**2+(p/x)/(x+(p/x))))+(x**2-p)/(p-p/(x**2-(p/x)/(x-(p/x))))
display("Expresion",exp2)
display("Minima expresion",exp2.together())
```

```python
#Derivar y reemplazar
exp3=sp.tan(x)**(p/x)
display("Expresion",exp3)
display("Derivada",exp3.diff())
display("Para x=pi/4 entonces:",exp3.diff().subs(x,sp.pi/4))
```

```python
#Integrar
exp4=(1+sp.log(x))*sp.sqrt(1+(x*sp.log(x))**2)
display("Expresion",exp4)
display("Integral",sp.integrate(exp4,(x,1,p+1)))
```

```python
#Grafica
#1)
fig, ax3d=plt.subplots(1,1,figsize=(8,6),subplot_kw={"projection":"3d"})
x=y=np.linspace(-np.pi/2,np.pi/2,500)
X,Y=np.meshgrid(x,y)
ax3d.set_title("i)")
ax3d.set_xlabel("$X$",fontsize=13)
ax3d.set_ylabel("$Y$",fontsize=13)
ax3d.set_zlabel("$Z$",fontsize=13)
Z=np.cos((X**2+Y**2)/4*p)
graph=ax3d.plot_surface(X,Y,Z,color="#E8FF33")
```

```python
#2)
fig2, ax3d2=plt.subplots(1,1,figsize=(8,6),subplot_kw={"projection":"3d"})
x2=y2=np.linspace(-5,5,50)
X2,Y2=np.meshgrid(x2,y2)
ax3d2.set_title("i)")
ax3d2.set_xlabel("$X$",fontsize=13)
ax3d2.set_ylabel("$Y$",fontsize=13)
ax3d2.set_zlabel("$Z$",fontsize=13)
Z2=np.log(abs(Y2-X2**2-p))
graph2=ax3d2.plot_surface(X2,Y2,Z2,color="#33A6FF")
```
