```python
import pandas as pd

def menu():
  print('------MENU------')
  print('1) Suma telescopica.')
  print('2) Fallas.')
  print('3) Salir')
  return int(input())

def sumTe(n):
  #Documento 1018506713 ultimo digito: 3
  a=3
  s=0
  for i in range(n):    
    s=s+((a*(i+1)+2*a+1)/((i+1)**2+3*(i+1)+2))
  print('Resultado: ',s)

def fallas():
  fallas=pd.read_csv('/content/drive/MyDrive/Clases/Metodos Numericos/corte 2/quiz 1/Fallas.csv')
  noApru=[]
  nombreEst=[]
  porcFallas=[]
  totalHorasClase=fallas.shape[1]*2
  porcPerdi=totalHorasClase*0.15
  for i in range(len(fallas.iloc[:])):
    s=0
    for j in range(len(fallas.iloc[i])):
      s=s+fallas.iloc[i,j]
    if s>=porcPerdi:
      nombreEst.append('Humano '+str(i+1))
      noApru.append(s)
      porcFallas.append(str(round((100*s)/totalHorasClase,2))+'%')    
  fallasPer=pd.DataFrame()
  fallasPer['Nombre']=nombreEst
  fallasPer['# Fallas']=noApru
  fallasPer['% Fallas']=porcFallas
  print('---Se pierde con un 15% o ',round(porcPerdi,1),' de fallas---\n',fallasPer)

bandera=True
while bandera==True:
  try:
    opc=menu()
    if opc==1:
      sumTe(int(input('Digite Cuantas veces desea realizar la suma')))
    elif opc==2:
      fallas()
    elif opc==3:
      b=False
      while b==False:
        s=str(input('¿desea salir? si/no: ')).upper()      
        if s=='SI'or s=='S':
          print('Salir')
          b=True
          bandera=False
        elif s=='NO' or s=='N':
          b=True
          bandera=True
        else:        
          print("!Esta opcion no existe¡")
    else:
      print("!Esta opcion no existe¡")
  except ValueError:
    print("!Esta opcion no existe¡")
```
